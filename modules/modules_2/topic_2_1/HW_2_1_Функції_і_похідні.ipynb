1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
# Код на Python без бібліотек для вирахування похідної
def f(x):
    return 10 * x - 5

def f_derivative(x):
    return 10

# Чисельне обчислення похідної
def numerical_derivative(func, x, h=1e-5):
    return (func(x + h) - func(x)) / h

# Обчислення похідної від 1 до 10 і підтвердимо, що дійсно вона постійна
for x in range(1, 11):
    print(f'f\'({x}) = {f_derivative(x)} (numerical: {numerical_derivative(f, x)})')
```

```python
import matplotlib.pyplot as plt

# Діапазон значень x від 0 до 10 з кроком 0.1
x_values = [i * 0.1 for i in range(101)]  # Генеруємо 101 значення в межах [0, 10]

# Обчислення значень f(x) та f'(x)
y_values = [f(x) for x in x_values]
y_derivative_values = [f_derivative(x) for x in x_values]

# Побудова графіків
plt.figure(figsize=(10, 6))

# Графік функції f(x)
plt.plot(x_values, y_values, label="f(x) = 10x - 5", color="blue", linewidth=2)

# Графік похідної f'(x)
plt.plot(x_values, y_derivative_values, label="f'(x) = 10", color="red", linestyle="--", linewidth=2)

# Додаткові налаштування
plt.title("Графіки функції f(x) та її похідної f'(x)")
plt.xlabel("x (години навчання)")
plt.ylabel("Оцінка")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")  # Лінія осі y=0
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")  # Лінія осі x=0
plt.legend()
plt.grid(True)

# Відображення графіків
plt.show()
```

**Висновок**: Бачимо, що усі значення похідної стрімлять до 10, що відповідає тому, що функція лінійна. Тобто оцінка зростає на 10 балів за кожну годину навчання.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Оголошуємо змінну x для SymPy
x = sp.Symbol('x')

# Функції
func1 = sp.sin(x)
func2 = sp.ln(x)

# Обчислення похідних
derivative1 = sp.diff(func1, x)
derivative2 = sp.diff(func2, x)

# Виведення результатів
print(f"Похідна від sin(x): {derivative1}")
print(f"Похідна від ln(x): {derivative2}")

# Конвертація функцій та похідних в Python-функції
f1 = sp.lambdify(x, func1, modules='numpy')
f2 = sp.lambdify(x, func2, modules='numpy')
df1 = sp.lambdify(x, derivative1, modules='numpy')
df2 = sp.lambdify(x, derivative2, modules='numpy')

# Генерація значень x
x_vals = np.linspace(0.1, 10, 100)  # Для ln(x), уникаємо x=0, щоб не ділити на нуль.

# Обчислення значень функцій та похідних
y1 = f1(x_vals)
y2 = f2(x_vals)
dy1 = df1(x_vals)
dy2 = df2(x_vals)

# Побудова графіків
plt.figure(figsize=(12, 8))

# Графік sin(x) та його похідної
plt.subplot(2, 1, 1)
plt.plot(x_vals, y1, label="sin(x)", color="blue")
plt.plot(x_vals, dy1, label="cos(x)", color="red", linestyle="--")
plt.title("Графік sin(x) та його похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

# Графік ln(x) та його похідної
plt.subplot(2, 1, 2)
plt.plot(x_vals, y2, label="ln(x)", color="blue")
plt.plot(x_vals, dy2, label="1/x", color="red", linestyle="--")
plt.title("Графік ln(x) та його похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

# Відображення графіків
plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Оголошуємо змінну x для SymPy
x = sp.Symbol('x')
f = x**2 - 5*x + 1

# Обчислення похідної
f_prime = sp.diff(f, x)
print(f"Похідна функції: {f_prime}")

# Перетворення функцій у звичайні для обчислень
f_lambdified = sp.lambdify(x, f, 'numpy')
f_prime_lambdified = sp.lambdify(x, f_prime, 'numpy')

# Генерація значень x
x_vals = np.linspace(0, 100, 100)

# Обчислення значень функцій та похідних
y = f_lambdified(x_vals)
y_prime = f_prime_lambdified(x_vals)

# Побудова графіків
plt.figure(figsize=(10, 6))

# Графік функції та її похідної
plt.plot(x_vals, y, label="f(x) = x^2 - 5x + 1", color="blue")
plt.plot(x_vals, y_prime, label="f'(x) = 2x - 5", color="red", linestyle="--")
plt.title("Графік функції загальних витрат і її похідної")
plt.xlabel("x (кількість одиниць)")
plt.ylabel("Загальні витрати")
plt.legend()
plt.grid(True)

# Відображення графіків
plt.show()
```
